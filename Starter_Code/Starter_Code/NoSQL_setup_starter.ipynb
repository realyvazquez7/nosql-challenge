## Part 2: Update the Database

1. An exciting new halal restaurant just opened in Greenwich, but hasn't been rated yet. The magazine has asked you to include it in your analysis. Add the following restaurant "Penang Flavours" to the database.

```python
# Create a dictionary for the new restaurant data
new_restaurant ={
    "BusinessName":"Penang Flavours",
    "BusinessType":"Restaurant/Cafe/Canteen",
    "BusinessTypeID":"",
    "AddressLine1":"Penang Flavours",
    "AddressLine2":"146A Plumstead Rd",
    "AddressLine3":"London",
    "AddressLine4":"",
    "PostCode":"SE18 7DY",
    "Phone":"",
    "LocalAuthorityCode":"511",
    "LocalAuthorityName":"Greenwich",
    "LocalAuthorityWebSite":"http://www.royalgreenwich.gov.uk",
    "LocalAuthorityEmailAddress":"health@royalgreenwich.gov.uk",
    "scores":{
        "Hygiene":"",
        "Structural":"",
        "ConfidenceInManagement":""
    },
    "SchemeType":"FHRS",
    "geocode":{
        "longitude":"0.08384000",
        "latitude":"51.49014200"
    },
    "RightToReply":"",
    "Distance":4623.9723280747176,
    "NewRatingPending":True
}
```

```python
# Insert the new restaurant into the collection
result = db.establishments.insert_one(new_restaurant)

# Print the inserted document's ID
print("Inserted document ID:", result.inserted_id)
```

```python
# Check that the new restaurant was inserted
query = {'BusinessName': 'Penang Flavours'}

# Find the document corresponding to the new restaurant
new_restaurant = db.establishments.find_one(query)

# Check if the new restaurant document was found
if new_restaurant:
    print("New restaurant was inserted successfully:")
    pprint(new_restaurant)
else:
    print("New restaurant was not found. It may not have been inserted successfully.")
```

2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields
query = {"BusinessType": "Restaurant/Cafe/Canteen"}
fields = {"BusinessTypeID":1,"BusinessType":1}
results = establishments.find(query,fields)
for result in results:
    pprint(result)
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
# Update the new restaurant with the correct BusinessTypeID
establishments.update_one({"BusinessName":"Penang Flavours"}, { "$set": {'BusinessTypeID': 1} })
```

```python
# Confirm that the new restaurant was updated
query = {'BusinessName': 'Penang Flavours'}

# Find the document corresponding to the updated restaurant
updated_restaurant = establishments.find_one(query)

# Check if the updated restaurant document was found
if updated_restaurant:
    print("New restaurant was updated successfully:")
    pprint(updated_restaurant)
else:
    print("New restaurant was not found. It may not have been updated successfully.")
```

4. The magazine is not interested in any establishments in Dover, so check how many documents contain the Dover Local Authority. Then, remove any establishments within the Dover Local Authority from the database, and check the number of documents to ensure they were deleted.

```python
# Find how many documents have LocalAuthorityName as "Dover"
query = {'LocalAuthorityName': 'Dover'}

# Count the documents matching the query
count = establishments.count_documents(query)

# Print the count
print("Number of documents with LocalAuthorityName as 'Dover':", count)
```

```python
# Delete all documents where LocalAuthorityName is "Dover"
query = {'LocalAuthorityName': 'Dover'}

# Delete the documents matching the query
delete_result = establishments.delete_many(query)

# Print the number of documents deleted
pprint(delete_result)
```

```python
# Check if any remaining documents include Dover
query = {'LocalAuthorityName': 'Dover'}

# Count the documents matching the query
count = establishments.count_documents(query)

# Check if any remaining documents include "Dover"
if count == 0:
    print("No remaining documents include 'Dover' in the LocalAuthorityName field.")
else:
    print("There are remaining documents that include 'Dover' in the LocalAuthorityName field.")
```

```python
# Check that other documents remain with 'find_one'
remaining_document = establishments.find_one()

# Check if a document was found
if remaining_document:
    print("At least one document remains in the collection after deleting documents with 'LocalAuthorityName' as 'Dover':")
    pprint(remaining_document)
else:
    print("No documents found in the collection.")
```

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
establishments.update_many({}, [{"$set":{"geocode.longitude":{"$toDecimal":"$geocode.longitude"}}}])
establishments.update_many({}, [{"$set":{"geocode.longitude":{"$toDouble":"$geocode.longitude"}}}])
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
establishments.update_many({},[{"$set":{"geocode.latitude":{"$toDecimal":"$geocode.latitude"}}}])
establishments.update_many({},[{"$set":{"geocode.latitude":{"$toDouble":"$geocode.latitude"}}}])
```

```python
# Check that the coordinates and rating value are now numbers
sample_document = establishments.find_one()

# Print the data types of the latitude, longitude, and RatingValue fields
if sample_document:
    print("Data type of latitude:", type(sample_document['geocode']['latitude']))
    print("Data type of longitude:", type(sample_document['geocode']['longitude']))
    print("Data type of RatingValue:", type(sample_document['RatingValue']))
else:
    print("No documents found in the collection.")
```
